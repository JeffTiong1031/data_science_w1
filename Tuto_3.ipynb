```python
import numpy as np

# 1. Create a numpy array of size 2 x 2
arr_2x2 = np.array([[1, 2],
                    [3, 4]])
print("2 x 2 array:\n", arr_2x2)
```

```python
# 2. Create a random numpy array of size 3 x 4
rand_3x4 = np.random.rand(3, 4)
print("\nRandom 3 x 4 array:\n", rand_3x4)
```

```python
# 3. Create a numpy array of ones and a numpy array of zeros
ones_arr = np.ones((3, 3))
zeros_arr = np.zeros((3, 3))
print("\nArray of ones:\n", ones_arr)
print("\nArray of zeros:\n", zeros_arr)
```

```python
# 4. Create a random numpy array of size 6 x 3 and transpose it
rand_6x3 = np.random.rand(6, 3)
transpose_6x3 = rand_6x3.T
print("\nRandom 6 x 3 array:\n", rand_6x3)
print("\nTransposed array (3 x 6):\n", transpose_6x3)
```

```python
# 5. Create two random numpy arrays of size n x p and p x m and then compute the dot product
n, p, m = 2, 3, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)
dot_product = np.dot(A, B)

print("\nMatrix A (",n,"x",p,"):\n", A)
print("\nMatrix B (",p,"x",m,"):\n", B)
print("\nDot Product (A x B):\n", dot_product)
```
